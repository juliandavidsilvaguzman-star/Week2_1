```python
# ============================
# Perceptrón Básico - Semana 2
# ============================

import numpy as np

# 1. Definir entradas, pesos y bias
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])   # Entradas (ejemplo: compuerta lógica AND)

w = np.array([0.5, 0.5]) # Pesos iniciales
b = -0.7                 # Sesgo (bias)

# 2. Función para calcular puntaje z
def calcular_z(x, w, b):
    return np.dot(x, w) + b

# 3. Función de activación (umbral)
def activacion(z):
    return 1 if z >= 0 else 0

# 4. Evaluar el perceptrón en todas las entradas
resultados = []
for x in X:
    z = calcular_z(x, w, b)
    y = activacion(z)
    resultados.append((x, z, y))

# 5. Mostrar resultados
print("Entrada | z | Salida")
for entrada, z, salida in resultados:
    print(f"{entrada} | {z:.2f} | {salida}")
```
